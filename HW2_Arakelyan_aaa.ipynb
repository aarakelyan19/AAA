```python
import csv
file = open("/Users/arturarakelan/Downloads/Corp_Summary-3.csv")
csvreader = csv.reader(file)
header = next(csvreader)
print(header)
rows = []
for row in csvreader:
    rows.append(row)
print(rows)
```

```python
lists_data = []

for i in range(len(rows)):
    a = rows[i][0].split(';')
    lists_data.append(a)
```

# 1 пункт

```python
def first_point(data:list) -> str:

    departaments = []

    for i in range(len(data)):
        dep = data[i][1]
        if dep in departaments:
            continue
        else:
            departaments.append(dep)
    teams = []

    for i in range(len(departaments)):
        teams_in_dep = []
        for j in range(len(data)):
            if departaments[i] == data[j][1]:
                team = data[j][2]
                if team in teams_in_dep:
                    continue
                else:
                    teams_in_dep.append(team)
        teams.append(teams_in_dep)
        
    for i in range(len(departaments)):
        args = teams[i]
        print(f'Департамент: {departaments[i]} -> команды: {*args, } ')
```

```python
first_point(lists_data)
```

# 2 пункт

```python
def second_point(data:list) -> str:
    global departaments
    global number_of_workers
    global mins
    global maxs
    global avgs

    departaments = []
    
    for i in range(len(data)):
        dep = data[i][1]
        if dep in departaments:
            continue
        else:
            departaments.append(dep)
            
    number_of_workers = []

    for i in range(len(departaments)):
        s = 0 
        for j in range(len(data)):
            if departaments[i] == data[j][1]:
                s += 1
        number_of_workers.append(s)
        
        
    mins = []

    for i in range(len(departaments)):
        min_salary = 100000000000
        for j in range(len(data)):
            if departaments[i] == data[j][1]:
                if int(data[j][5]) < min_salary:
                    min_salary = int(data[j][5])
                else:
                    continue
        mins.append(min_salary)
        
    maxs = []

    for i in range(len(departaments)):
        max_salary = 0
        for j in range(len(data)):
            if departaments[i] == data[j][1]:
                if int(data[j][5]) > max_salary:
                    max_salary = int(data[j][5])
                else:
                    continue
        maxs.append(max_salary)
    
    avgs = []

    for i in range(len(departaments)):
        avg = 0
        l = 0
        for j in range(len(data)):
            if departaments[i] == data[j][1]:
                avg = (avg*l+int(data[j][5]))/(l+1)
                l += 1
        avgs.append(avg)
    
    for i in range(len(departaments)):
        print(f'Департамент: {departaments[i]}, численность: {number_of_workers[i]}, вилка: {mins[i]} - {maxs[i]}, средняя зп: {avgs[i]}')
```

```python
second_point(lists_data)
```

# 3 пункт

```python
# csv header
def third_point(data:list):
    
    second_point(data)
    
    fieldnames = ['департамент', 'численность', 'вилка', 'средняя зп']
    rows = []
    
    for i in range(len(departaments)):
        row = {'департамент': departaments[i],
        'численность':number_of_workers[i],
        'вилка': f'{mins[i]} - {maxs[i]}',
        'средняя зп': avgs[i]}
        rows.append(row)

    with open('third_point_hw2_aaa.csv', 'w', encoding='UTF8', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(rows)
```

```python
third_point(lists_data)
```

```python
point = input("Какой пункт?")
print(point)

while True:
    if point == '1':
        first_point(lists_data)
        break
    elif point == '2':
        second_point(lists_data)
        break
    elif point == '3':
        third_point(lists_data)
        break
    else:
        print('Такого пункта нет, давайте еще раз')
        point = input("Какой пункт?")
```
